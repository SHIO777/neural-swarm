```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
from scipy import signal
from random import shuffle
```

```python
# Convert quaternion to rotation matrix
def rotation_matrix(quat):
    rot_mat = np.ones([3,3])
    a = quat[0]**2
    b = quat[1]**2
    c = quat[2]**2
    d = quat[3]**2
    e = quat[0]*quat[1]
    f = quat[0]*quat[2]
    g = quat[0]*quat[3]
    h = quat[1]*quat[2]
    i = quat[1]*quat[3]
    j = quat[2]*quat[3]
    rot_mat[0,0] = a - b - c + d
    rot_mat[0,1] = 2 * (e - j)
    rot_mat[0,2] = 2 * (f + i)
    rot_mat[1,0] = 2 * (e + j)
    rot_mat[1,1] = -a + b - c + d
    rot_mat[1,2] = 2 * (h - g)
    rot_mat[2,0] = 2 * (f - i)
    rot_mat[2,1] = 2 * (h + g)
    rot_mat[2,2] = -a - b + c + d
     
    return rot_mat
```

```python
# Convert quaternion to Euler angle
def qua2euler(qua):
    euler = np.zeros(3)
    q0 = qua[3]
    q1 = qua[0]
    q2 = qua[1]
    q3 = qua[2]
    euler[0] = np.degrees(np.arctan2(2*(q0*q1+q2*q3), 1-2*(q1**2+q2**2)))
    euler[1] = np.degrees(np.arcsin(2*(q0*q2-q3*q1)))
    euler[2] = np.degrees(np.arctan2(2*(q0*q3+q1*q2), 1-2*(q2**2+q3**2)))
    return euler
```

# Preprocessing data

```python
def data_extraction(filename):    
    raw = pd.read_csv(filename)
    secs = np.copy(raw.iloc[:, 0])
    bias = np.copy(secs[0])
    time = (secs - bias) / 1e6
    pos = np.copy(raw.iloc[:, 1:4])
    vel = np.copy(raw.iloc[:, 4:7])
    acc = np.copy(raw.iloc[:, 7:10])
    qua = np.copy(raw.iloc[:, 10:14])
    thr = np.copy(raw.iloc[:, 14:18])
    pwm = np.copy(raw.iloc[:, 18:22])
    vol = np.copy(raw.iloc[:, 22]) / 4.2
    thr_des = np.copy(raw.iloc[:, 23])
    tau_u = np.copy(raw.iloc[:, 24:27])
    omega = np.copy(raw.iloc[:, 27:])
    
    Data = {'time': time, 'pos': pos, 'vel': vel, 'acc': acc, 'qua': qua, 'thr': thr, \
            'pwm': pwm, 'vol': vol, 'thr_des': thr_des, 'tau_u': tau_u, 'omega': omega}
    
    return Data
```

**Interpolation**

```python
from scipy import interpolate
from scipy.interpolate import interp1d
```

```python
# cubic (or linear) interpolation
def cubic(x, y, xnew, kind='linear'):
    f = interp1d(x, y, kind=kind)
    return f(xnew)
    
def interpolation_cubic(t0, t1, Data, ss, ee):
    time = np.linspace(t0, t1, 100 * (t1 - t0) + 1)
    pos = np.zeros((time.shape[0], 3))
    qua = np.zeros((time.shape[0], 4))
    pwm = np.zeros((time.shape[0], 4))
    vel = np.zeros((time.shape[0], 3))
    vel_num = np.zeros((time.shape[0], 3))
    acc_num = np.zeros((time.shape[0], 3))
    vol = np.zeros((time.shape[0]))
    acc_imu = np.zeros((time.shape[0], 3))
    tau_u = np.zeros((time.shape[0], 3))
    omega = np.zeros((time.shape[0], 3))
    omega_dot = np.zeros((time.shape[0], 3))
    euler = np.zeros((time.shape[0], 3))
    acc_filter = np.zeros((time.shape[0], 3))
    acc_smooth = np.zeros((time.shape[0], 3))
    
    x = Data['time'][ss:ee]
    for i in range(3):
        pos[:, i] = cubic(x, Data['pos'][ss:ee, i], time)
        vel[:, i] = cubic(x, Data['vel'][ss:ee, i], time)
        acc_imu[:, i] = cubic(x, Data['acc'][ss:ee, i], time)
        tau_u[:, i] = cubic(x, Data['tau_u'][ss:ee, i], time)
        omega[:, i] = cubic(x, Data['omega'][ss:ee, i], time)
 
    for i in range(3):
        acc_num[2:-2,i] = (-vel[4:,i] + 8 * vel[3:-1,i] - 8 * vel[1:-3,i] + vel[:-4,i]) / 12 * 100
        vel_num[2:-2,i] = (-pos[4:,i] + 8 * pos[3:-1,i] - 8 * pos[1:-3,i] + pos[:-4,i]) / 12 * 100        
        omega_dot[2:-2,i] = (-omega[4:,i] + 8 * omega[3:-1,i] - 8 * omega[1:-3,i] + omega[:-4,i]) / 12 * 100
    
    for i in range(4):
        qua[:, i] = cubic(x, Data['qua'][ss:ee, i], time)
        pwm[:, i] = cubic(x, Data['pwm'][ss:ee, i], time)
    vol[:] = cubic(x, Data['vol'][ss:ee], time)
    
    for j in range(time.shape[0]):
        euler[j, :] = qua2euler(qua[j, :])
    
    # Filter on acc
    b, a = signal.butter(1, 0.1)
    for i in range(3):
        acc_filter[:, i] = signal.filtfilt(b, a, acc_num[:, i])
    
    # Moving average smoothing
    n = 5
    l = int((n-1) / 2)
    for i in range(3):
        for j in range(n):
            if j == n-1:
                temp = acc_num[j:, i]
            else:
                temp = acc_num[j:-(n-1-j), i]
            acc_smooth[l:-l, i] = acc_smooth[l:-l, i] + temp
        acc_smooth[l:-l, i] = acc_smooth[l:-l, i] / n
        
    Data_int = {'time': time, 'pos': pos, 'vel': vel, 'acc_imu': acc_imu, 'vel_num': vel_num, \
                'qua': qua, 'pwm': pwm, 'vol': vol, 'acc_num': acc_num, 'euler': euler, \
               'tau_u': tau_u, 'omega': omega, 'omega_dot': omega_dot, 'acc_filter': acc_filter, 'acc_smooth': acc_smooth}
    return Data_int
```

**Data merge**

```python
def merge(Data1, Data2):
    Data = {'time': np.concatenate((Data1['time'], Data2['time'] + Data1['time'][-1] + 0.01))}
    for topic in Data1:
        if topic != 'time':
            Data[topic] = np.concatenate((Data1[topic], Data2[topic]))
    return Data
```

```python
def Merge(Data_14_list):
    Data_14 = merge(Data_14_list[0], Data_14_list[1])
    if len(Data_14_list) > 2:
        for i in range(2, len(Data_14_list)):
            Data_14 = merge(Data_14, Data_14_list[i])
    return Data_14
```

# Get $F_a$ and $\tau_a$ data

**Get Fa from new system id and numerical diff**

```python
def Fa(Data, m, g, p_00, p_10, p_01, p_20, p_11):
    R = np.zeros([Data['time'].shape[0], 3, 3])
    for i in range(Data['time'].shape[0]):
        R[i, :, :] = rotation_matrix(Data['qua'][i, :])
        
    force_pwm_1 = p_00 + p_10 * Data['pwm'][:, 0] + p_01 * Data['vol'] + p_20 * Data['pwm'][:, 0]**2 + p_11 * Data['vol'] * Data['pwm'][:, 0]
    force_pwm_2 = p_00 + p_10 * Data['pwm'][:, 1] + p_01 * Data['vol'] + p_20 * Data['pwm'][:, 1]**2 + p_11 * Data['vol'] * Data['pwm'][:, 1]
    force_pwm_3 = p_00 + p_10 * Data['pwm'][:, 2] + p_01 * Data['vol'] + p_20 * Data['pwm'][:, 2]**2 + p_11 * Data['vol'] * Data['pwm'][:, 2]
    force_pwm_4 = p_00 + p_10 * Data['pwm'][:, 3] + p_01 * Data['vol'] + p_20 * Data['pwm'][:, 3]**2 + p_11 * Data['vol'] * Data['pwm'][:, 3]
    thrust_pwm = force_pwm_1 + force_pwm_2 + force_pwm_3 + force_pwm_4 # gram
    
    Fa = np.zeros([Data['time'].shape[0], 3])
    Fa_num = np.zeros([Data['time'].shape[0], 3])
    Fa_filter = np.zeros([Data['time'].shape[0], 3])
    Fa_smooth = np.zeros([Data['time'].shape[0], 3])
    force_world = np.zeros([Data['time'].shape[0], 3])
    for i in range(Data['time'].shape[0]):
        Fa[i, :] = m * Data['acc_imu'][i, :] / 1000 - thrust_pwm[i] / 1000 * g * R[i, :, 2] # Newton
        Fa_num[i, :] = m * np.array([0, 0, g]) / 1000 + m * Data['acc_num'][i, :] / 1000 - thrust_pwm[i] / 1000 * g * R[i, :, 2] # Newton
        Fa_filter[i, :] = m * np.array([0, 0, g]) / 1000 + m * Data['acc_filter'][i, :] / 1000 - thrust_pwm[i] / 1000 * g * R[i, :, 2] # Newton
        Fa_smooth[i, :] = m * np.array([0, 0, g]) / 1000 + m * Data['acc_smooth'][i, :] / 1000 - thrust_pwm[i] / 1000 * g * R[i, :, 2] # Newton
        force_world[i, :] = thrust_pwm[i] / 1000 * g * R[i, :, 2] # Newton
    
    Data['fa_imu'] = Fa
    Data['fa_num'] = Fa_num
    Data['fa_filter'] = Fa_filter
    Data['fa_smooth'] = Fa_smooth
    
    return Data
```

**Data normalization**

```python
def get_data(D1, D2, GE=True):
    typ = 'fa_filter'
    if D2 is None:
        L = D1['time'].shape[0]
        data_input = np.zeros([L, 10])
        data_output = np.zeros([L, 3])
        data_input[:L, 6] = 0 - D1['pos'][:, 2]
        data_input[:L, 7:] = 0 - D1['vel'][:, :]
        data_output[:L, :] = D1[typ] / g * 1000 # gram
    else:
        L = D1['time'].shape[0]
        data_input = np.zeros([2*L, 10])
        data_output = np.zeros([2*L, 3])
        data_input[:L, :3] = D2['pos'] - D1['pos']
        data_input[:L, 3:6] = D2['vel'] - D1['vel']
        if GE is True:
            data_input[:L, 6] = 0 - D1['pos'][:, 2]
            data_input[:L, 7:] = 0 - D1['vel'][:, :]
        data_output[:L, :] = D1[typ] / g * 1000 # gram

        data_input[L:, :3] = D1['pos'] - D2['pos']
        data_input[L:, 3:6] = D1['vel'] - D2['vel']
        if GE is True:
            data_input[L:, 6] = 0 - D2['pos'][:, 2]
            data_input[L:, 7:] = 0 - D2['vel'][:, :]
        data_output[L:, :] = D2[typ] / g * 1000 # gram
        
    print(data_input.shape, data_output.shape)
    
    return data_input, data_output
```

```python
from torch.utils.data import Dataset, DataLoader

class MyDataset(Dataset):

    def __init__(self, inputs, outputs):
        self.inputs = inputs
        self.outputs = outputs

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):
        Input = self.inputs[idx,]
        output = self.outputs[idx,]
        sample = {'input': Input, 'output': output}

        return sample
```

```python
def set_generate(Data_input, Data_output):
    trainset = MyDataset(Data_input, Data_output)
    trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True, num_workers=2)
    return trainset, trainloader
```

```python
trainset_2, trainloader_2 = set_generate(Data_input_two, Data_output_two)
trainset_1, trainloader_1 = set_generate(Data_input_one, Data_output_one)
trainset_2_free, trainloader_2_free = set_generate(Data_input_two_free, Data_output_two_free)
```

**Neural net and optimization**

```python
class phi_Net(nn.Module):
    def __init__(self):
        super(phi_Net, self).__init__()
        self.fc1 = nn.Linear(6, 25)
        self.fc2 = nn.Linear(25, 40)
        self.fc3 = nn.Linear(40, 40)
        self.fc4 = nn.Linear(40, 40)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = self.fc4(x)
        return x

class phi2_Net(nn.Module):
    def __init__(self):
        super(phi2_Net, self).__init__()
        self.fc1 = nn.Linear(4, 25)
        self.fc2 = nn.Linear(25, 40)
        self.fc3 = nn.Linear(40, 40)
        self.fc4 = nn.Linear(40, 40)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = self.fc4(x)
        return x

class phi3_Net(nn.Module):
    def __init__(self):
        super(phi3_Net, self).__init__()
        self.fc1 = nn.Linear(4, 25)
        self.fc2 = nn.Linear(25, 40)
        self.fc3 = nn.Linear(40, 40)
        self.fc4 = nn.Linear(40, 1)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = self.fc4(x)
        return x
    
class rho_Net(nn.Module):
    def __init__(self):
        super(rho_Net, self).__init__()
        self.fc1 = nn.Linear(40, 40)
        self.fc2 = nn.Linear(40, 40)
        self.fc3 = nn.Linear(40, 40)
        self.fc4 = nn.Linear(40, 1)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = self.fc4(x)
        return x

phi_net = phi_Net()
phi2_net = phi2_Net()
phi3_net = phi3_Net()
rho_net = rho_Net()
```

```python
criterion = nn.MSELoss()
optimizer_phi = optim.Adam(phi_net.parameters(), lr=1e-3)
optimizer_phi2 = optim.Adam(phi2_net.parameters(), lr=1e-3)
optimizer_phi3 = optim.Adam(phi3_net.parameters(), lr=1e-3)
optimizer_rho = optim.Adam(rho_net.parameters(), lr=1e-3)
```

**Training**

```python
# Loss before training
train_loss_2 = 0
train_loss_1 = 0
train_loss_2_free = 0
train_loss_1_phi3 = 0
with torch.no_grad():
    for i in range(len(trainset_2)):
        inputs = trainset_2[i]['input'] 
        label = trainset_2[i]['output']
        loss = criterion(rho_net(phi_net(inputs[:6]) + phi2_net(inputs[6:])), label)
        train_loss_2 += loss.item()
    for i in range(len(trainset_1)):
        inputs = trainset_1[i]['input'] 
        label = trainset_1[i]['output']
        loss = criterion(rho_net(phi2_net(inputs[6:])), label)
        train_loss_1 += loss.item()
        loss = criterion(phi3_net(inputs[6:]), label)
        train_loss_1_phi3 += loss.item()
    for i in range(len(trainset_2_free)):
        inputs = trainset_2_free[i]['input'] 
        label = trainset_2_free[i]['output']
        loss = criterion(rho_net(phi_net(inputs[:6])), label)
        train_loss_2_free += loss.item()
        
print(train_loss_2 / len(trainset_2))
print(train_loss_1 / len(trainset_1))
print(train_loss_2_free / len(trainset_2_free))
print(train_loss_1_phi3 / len(trainset_1))
```

```python
Loss_sn = []

# Loss b4 training
with torch.no_grad():
    running_loss = 0.0
    for i, data in enumerate(trainloader_2, 0):
        # pass
        # get the inputs
        inputs = data['input']
        labels = data['output']
        outputs = rho_net(phi_net(inputs[:, :6]) + phi2_net(inputs[:, 6:]))
        loss = criterion(outputs, labels)
        running_loss += loss.item()
    for i, data in enumerate(trainloader_1, 0):
        # get the inputs
        inputs = data['input']
        labels = data['output']
        outputs = rho_net(phi2_net(inputs[:, 6:]))
        #outputs = phi3_net(inputs[:, 6:])
        loss = criterion(outputs, labels)
        running_loss += 2 * loss.item()
    for i, data in enumerate(trainloader_2_free, 0):
        # pass
        # get the inputs
        inputs = data['input']
        labels = data['output']
        outputs = rho_net(phi_net(inputs[:, :6]))
        loss = criterion(outputs, labels)
        running_loss += loss.item()
    Loss_sn.append(running_loss)
    print('Initial loss:' + str(running_loss))
    
for epoch in range(150):  # loop over the dataset multiple times
    running_loss = 0.0
    mixed = []
    count_2 = 0
    count_1 = 0
    count_2_free = 0
    for i, data in enumerate(trainloader_2, 0):
        # pass
        mixed.append(data)
    for i, data in enumerate(trainloader_1, 0):
        mixed.append(data)
        mixed.append(data)
    for i, data in enumerate(trainloader_2_free, 0):
        # pass
        mixed.append(data)
        
    shuffle(mixed)
    
    for data in mixed:
        # get the inputs
        inputs = data['input']
        labels = data['output']

        # zero the parameter gradients
        optimizer_phi.zero_grad()
        optimizer_phi2.zero_grad() 
        # optimizer_phi3.zero_grad() 
        optimizer_rho.zero_grad()

        # forward + backward + optimize
        temp = torch.sum(inputs[:, :6]).item()
        temp2 = torch.sum(inputs[:, 6:]).item()
        if temp == 0:
            outputs = rho_net(phi2_net(inputs[:, 6:]))
            # outputs = phi3_net(inputs[:, 6:])
            count_1 += 1
        elif temp2 == 0:
            outputs = rho_net(phi_net(inputs[:, :6]))
            count_2_free += 1
        else:
            outputs = rho_net(phi_net(inputs[:, :6]) + phi2_net(inputs[:, 6:]))
            count_2 += 1
        
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer_phi.step()
        optimizer_phi2.step()
        # optimizer_phi3.step()
        optimizer_rho.step()

        # Lip
        for param in phi_net.parameters():
            M = param.detach().numpy()
            if M.ndim > 1:
                s = np.linalg.norm(M, 2)
                if s > 3.5:
                    param.data = param / s * 3.5
        
        for param in phi2_net.parameters():
            M = param.detach().numpy()
            if M.ndim > 1:
                s = np.linalg.norm(M, 2)
                if s > 3.5:
                    param.data = param / s * 3.5              
                    
        for param in rho_net.parameters():
            M = param.detach().numpy()
            if M.ndim > 1:
                s = np.linalg.norm(M, 2)
                if s > 3.5:
                    param.data = param / s * 3.5
        '''
        for param in phi3_net.parameters():
            M = param.detach().numpy()
            if M.ndim > 1:
                s = np.linalg.norm(M, 2)
                if s > 4.0:
                    param.data = param / s * 4.0      
        '''
        
        running_loss += loss.item()

    if count_2 != len(trainloader_2) or count_1 != 2*len(trainloader_1) or count_2_free != len(trainloader_2_free):
    # if count_1 != len(trainloader_1):
        print('something goes wrong!')
    Loss_sn.append(running_loss)
    if epoch % 10 == 0:
        print('[%d] loss: %.3f' % (epoch + 1, running_loss))

print('Finished Training')
plt.plot(Loss_sn)
plt.show()
```

```python
# Loss before training
train_loss_2 = 0
train_loss_1 = 0
train_loss_2_free = 0
train_loss_1_phi3 = 0
with torch.no_grad():
    for i in range(len(trainset_2)):
        inputs = trainset_2[i]['input'] 
        label = trainset_2[i]['output']
        loss = criterion(rho_net(phi_net(inputs[:6]) + phi2_net(inputs[6:])), label)
        train_loss_2 += loss.item()
    for i in range(len(trainset_1)):
        inputs = trainset_1[i]['input'] 
        label = trainset_1[i]['output']
        loss = criterion(rho_net(phi2_net(inputs[6:])), label)
        train_loss_1 += loss.item()
        loss = criterion(phi3_net(inputs[6:]), label)
        train_loss_1_phi3 += loss.item()
    for i in range(len(trainset_2_free)):
        inputs = trainset_2_free[i]['input'] 
        label = trainset_2_free[i]['output']
        loss = criterion(rho_net(phi_net(inputs[:6])), label)
        train_loss_2_free += loss.item()
        
print(train_loss_2 / len(trainset_2))
print(train_loss_1 / len(trainset_1))
print(train_loss_2_free / len(trainset_2_free))
print(train_loss_1_phi3 / len(trainset_1))
```

```python
def heatmap(phi_net, phi2_net, rho_net, phi3_net, vz_1=0.0, x_2=0, y_2=0, z_2=0.5, vx_2=0, vy_2=0, vz_2=0):
    z_min = 0.05
    z_max = 1.0
    y_min = -0.5
    y_max = 0.5
    
    z_length = int((z_max-z_min)*50) + 1
    y_length = int((y_max-y_min)*50) + 1
    
    z = np.linspace(z_max, z_min, z_length)
    y = np.linspace(y_min, y_max, y_length)

    vz_min = -0.5
    vz_max = 0.5
    vy_min = -0.5
    vy_max = 0.5
    
    vz_length = int((vz_max-vz_min)*50) + 1
    vy_length = int((vy_max-vy_min)*50) + 1
    
    vz = np.linspace(vz_max, vz_min, vz_length)
    vy = np.linspace(vy_min, vy_max, vy_length)
    
    fa_heatmap = np.zeros([1, z_length, y_length])
    fa_heatmap_ge = np.zeros([1, z_length, y_length])
    fa_heatmap_free = np.zeros([1, z_length, y_length])
    fa_heatmap_ge_v = np.zeros([1, z_length, vz_length])
    
    for j in range(y_length):
        for i in range(z_length):
            c = np.zeros([1, 10])
            c[0, 0] = x_2 - 0
            c[0, 1] = y_2 - y[j]
            c[0, 2] = z_2 - z[i]
            c[0, 3] = vx_2
            c[0, 4] = vy_2
            c[0, 5] = vz_2
            c[0, 6] = 0 - z[i]
            c[0, 9] = 0 - vz_1
            cc = torch.from_numpy(c)
            with torch.no_grad():
                fa_heatmap[0, i, j] = rho_net(phi_net(cc[:, :6])+phi2_net(cc[:, 6:]))[0, 0].item() # f_a_z
                fa_heatmap_ge[0, i, j] = rho_net(phi2_net(cc[:, 6:]))[0, 0].item() # f_a_z
                fa_heatmap_free[0, i, j] = rho_net(phi_net(cc[:, :6]))[0, 0].item()
    
    for j in range(vz_length):
        for i in range(z_length):
            c = np.zeros([1, 10])
            c[0, 6] = 0 - z[i]
            c[0, 9] = 0 - vz[j]
            cc = torch.from_numpy(c)
            with torch.no_grad():
                fa_heatmap_ge_v[0, i, j] = rho_net(phi2_net(cc[:, 6:]))[0, 0].item() # f_a_z
                # fa_heatmap_ge_v[0, i, j] = phi3_net(cc[:, 6:])[0, 0].item() # f_a_z
    
    plt.figure(figsize=(16,4))
    plt.subplot(1, 4, 1)
    plt.ylabel(r'$z$ (m)', fontsize = 15)
    plt.xlabel(r'$y$ (m)', fontsize = 15)
    plt.title(r'${F_a}_z$ (g)', fontsize = 15)
    plt.pcolor(y, z, fa_heatmap[0, :, :], cmap='Reds_r')
    plt.colorbar()
    plt.tick_params(labelsize = 13)
    plt.plot([y_2], [z_2], marker='*', markersize=10, color="black")
    
    plt.subplot(1, 4, 2)
    plt.ylabel(r'$z$ (m)', fontsize = 15)
    plt.xlabel(r'$y$ (m)', fontsize = 15)
    plt.title(r'${F_a}_z$ (g)', fontsize = 15)
    plt.pcolor(y, z, fa_heatmap_ge[0, :, :], cmap='Reds_r')
    plt.colorbar()
    plt.tick_params(labelsize = 13)
    plt.plot([y_2], [z_2], marker='*', markersize=10, color="black")
    
    plt.subplot(1, 4, 3)
    plt.ylabel(r'$z$ (m)', fontsize = 15)
    plt.xlabel(r'$y$ (m)', fontsize = 15)
    plt.title(r'${F_a}_z$ (g)', fontsize = 15)
    plt.pcolor(y, z, fa_heatmap_free[0, :, :], cmap='Reds_r')
    plt.colorbar()
    plt.tick_params(labelsize = 13)
    plt.plot([y_2], [z_2], marker='*', markersize=10, color="black")
    
    plt.subplot(1, 4, 4)
    plt.ylabel(r'$z$ (m)', fontsize = 15)
    plt.xlabel(r'$v_z$ (m)', fontsize = 15)
    plt.title(r'${F_a}_z$ (g)', fontsize = 15)
    plt.pcolor(vz, z, fa_heatmap_ge_v[0, :, :], cmap='Reds_r')
    plt.colorbar()
    plt.tick_params(labelsize = 13)
    plt.tight_layout()
    plt.show()
```
